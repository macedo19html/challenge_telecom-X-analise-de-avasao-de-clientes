# üìä Relat√≥rio de An√°lise de Churn ‚Äî Telecom X

## 1. Introdu√ß√£o

A **Telecom X** enfrenta um problema significativo de evas√£o de clientes, conhecido como **churn**, que impacta negativamente sua receita e crescimento.  
O objetivo desta an√°lise √© identificar os fatores que contribuem para o cancelamento dos contratos, utilizando t√©cnicas de **an√°lise explorat√≥ria de dados**.  
Compreender esses fatores permitir√° que a equipe de Data Science desenvolva modelos preditivos e estrat√©gias para **reduzir a evas√£o**.

---

## 2. Limpeza e Tratamento de Dados

Os dados foram obtidos de uma fonte online contendo informa√ß√µes detalhadas dos clientes, incluindo dados demogr√°ficos, servi√ßos contratados, faturamento e status de cancelamento.

As principais etapas de limpeza e tratamento foram:

- üîÑ **Desempacotamento** de colunas com dados aninhados (dicion√°rios) para transformar em um formato tabular.
- üßº **Preenchimento de valores faltantes** na coluna `Churn` com `'N√£o'`, garantindo a completude dos dados.
- üî¢ **Convers√£o de colunas categ√≥ricas** do formato `'Yes'/'No'` para valores bin√°rios `1/0` para facilitar an√°lises num√©ricas.
- üåç **Tradu√ß√£o das colunas** para o portugu√™s para melhorar a compreens√£o.
- ‚ûó **Cria√ß√£o da coluna `Contas_Diarias`**, que calcula o gasto m√©dio di√°rio a partir do faturamento mensal.

Esses passos garantiram um conjunto de dados **limpo e padronizado**, pronto para an√°lise.

- Clientes **idosos**, com **contratos mensais**, cuja  **forma de pagamento √© eletr√¥nico** e que **n√£o tem contrato de suporte t√©cnico** s√£o os mais propensos ao cancelamento.

## 5. Recomenda√ß√µes

- üéÅ Oferecer **planos e incentivos personalizados para clientes idosos**, como descontos ou suporte exclusivo.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
df = pd.read_json(url)
df.head()
```

```python
# Normalizar cada coluna com dicion√°rios
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

# Colunas restantes (como Churn)
df_other = df.drop(columns=['customer', 'phone', 'internet', 'account'])
```

```python
# Concatenar todas as partes
df = pd.concat([df_customer, df_phone, df_internet, df_account, df_other], axis=1)
```

```python
# Verificando valores ausentes por coluna
df.isnull().sum().sort_values(ascending=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Lista das vari√°veis categ√≥ricas
variaveis = [
    'Idoso', 'Suporte_Tecnico',
    'Contrato', 'Metodo_Pagamento'
]

# Lista para armazenar os dados
maiores_churn = []

# Traduzir 0/1 para Sim/N√£o em categorias booleanas
def traduzir_categoria(valor):
    if valor == 1:
        return 'Sim'
    elif valor == 0:
        return 'N√£o'
    return str(valor)

# Calcular a maior porcentagem de churn por vari√°vel
for var in variaveis:
    porcentagens = df.groupby(var)['Cancelou'].mean() * 100
    categoria_mais_churn = porcentagens.idxmax()
    maior_valor = porcentagens.max()
    categoria_legenda = f"{var} = {traduzir_categoria(categoria_mais_churn)}"
    maiores_churn.append({
        'Vari√°vel': var,
        'Categoria': categoria_mais_churn,
        'Categoria_Legenda': categoria_legenda,
        'Porcentagem_Churn': maior_valor
    })

# Criar DataFrame com os resultados
df_maiores = pd.DataFrame(maiores_churn).sort_values(by='Porcentagem_Churn', ascending=False)

# Obter categorias √∫nicas para a legenda
categorias_unicas = df_maiores['Categoria_Legenda'].unique()

# Gerar uma paleta de cores com a quantidade exata de categorias √∫nicas
cores_personalizadas = sns.color_palette("Set2", n_colors=len(categorias_unicas))

# Criar dicion√°rio de cor por categoria
cores_dict = dict(zip(categorias_unicas, cores_personalizadas))

# Plotar gr√°fico com as cores definidas
plt.figure(figsize=(12, 7))
plot = sns.barplot(
    data=df_maiores,
    y='Vari√°vel',
    x='Porcentagem_Churn',
    hue='Categoria_Legenda',
    palette=cores_dict
)

# Atualizar legenda com todas as categorias
handles, labels = plot.get_legend_handles_labels()
plot.legend(
    handles=handles,
    labels=labels,
    title='Categoria com maior churn',
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

# T√≠tulos e ajustes
plt.title('Maiores Porcentagens de Cancelamento por Categoria')
plt.xlabel('Porcentagem de Cancelamento (%)')
plt.ylabel('Vari√°vel')
plt.tight_layout()

plt.savefig("graficos/maiores_porcentagens_churn.png", dpi=300, bbox_inches='tight')


plt.show()
```
